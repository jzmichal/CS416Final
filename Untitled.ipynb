```python
import pandas as pd
```

```python
df = pd.read_csv('cars2017.csv')
```

```python
df.head()
```

```python
df["AvgCityandHwMPG"] = (df["AverageHighwayMPG"] + df["AverageCityMPG"]) / 2
```

```python
df.head()
```

```python
df = df.join(
    df.groupby('Make')["AvgCityandHwMPG"]
        .transform('mean')  # Calculate the mean for each group
        .rename("MakeAvgMPG")  # Rename columns 
)
```

```python
df.head()
```

```python
df = df.join(
    df.groupby('Make')["AvgCityandHwMPG"]
        .transform('min')  # Calculate the mean for each group
        .rename("MakeMinMPG")  # Rename columns 
)
```

```python
df.head()
```

```python
df.to_csv('cars2017.csv', index = True)
```
